# Hypothesis


*State and defend your hypotheses here.*

Animatronics at SeaWorld would not reduce engagement from visitors significantly. Not only would animatronics be less expensive and easier to care for, but visitors could also more safely and closely engage with the exhibits at theme parks. In the long term, this could even generate more interest in biology and robotics.
